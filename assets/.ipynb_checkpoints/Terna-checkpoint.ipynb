```python
import io
import os
import requests
import logging
import pandas as pd

logger = logging.getLogger('log')
logger.setLevel('INFO')

terna_path = 'terna'
os.makedirs(terna_path, exist_ok=True)

terna_url = 'http://download.terna.it/terna/0000/0{server_dir}/{file}.XLS'
```

```python
%%time
session = requests.session()
for i in range(600, 660):
    for j in range(0, 100):
        url = terna_url.format(server_dir=i, file=j)
        resp = session.head(url)
        #resp = session.get(url)
        if resp.status_code == 200:
            filesize = int(resp.headers['Content-Length'])/1024
            logger.info('%s/%s: %s', i, j, resp.status_code)
            if 150 < filesize < 600:
                logger.info('this one has the right size')
                resp = session.get(url)
                filename = '{server_dir}-{file}.XLS'.format(server_dir=i, file=j)
                filepath = os.path.join(terna_path, filename)
                with open(filepath, 'wb') as output_file:
                    for chunk in resp.iter_content(1024):
                        output_file.write(chunk)
        else: logger.info('%s/%s: %s', i, j, resp.status_code)
```
