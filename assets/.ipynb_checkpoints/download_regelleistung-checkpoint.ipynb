# Download script for regelleistung.net

The files are put on the server apparently continously. The URL to the file is made up of a counter that just increments by 1 with every new file. In order for the user to see which file timespan is associated with which range of the counter, the log output from previous download runs has been left in this notebook.

To download more data, just call the `download_regelleistung()` -Function with the start and the end-number that seem appropriate.

```python
import requests
import os
```

```python
path = 'Download/'
os.makedirs(path, exist_ok=True)
```

```python
def download_regelleistung(start, end):
    url_template = 'https://www.regelleistung.net/ext/tender/results/download/{count}'
    for i in range(start,end):
        full_url = url_template.format(count = i)
        resp = requests.get(full_url, stream = True)
        try:
            pre_original_filename = resp.headers['content-disposition'].split('filename=')[-1]
            original_filename = pre_original_filename.replace('"','').replace(';','')
            print(i, 'worked', original_filename)
        except KeyError:
            print(i, 'failed')
            continue
        full_filename = path + original_filename
        with open(full_filename, 'wb') as output_file:
            for chunk in resp.iter_content(1024):
                output_file.write(chunk)     
```

```python
download_regelleistung(1273, 1600)
```

```python
download_regelleistung(4600, 5210)
```

```python
download_regelleistung(5210, 5500)
```
